```python
# importing libraries
import pandas as pd
import json
```

```python
df = pd.read_csv("movies_metadata.csv")
df
```

```python
collection = []

for ele in df["belongs_to_collection"]:
    try:
        if pd.isnull(ele):
            collection.append(None)
        else:
            
            collection.append(json.loads(ele.replace("'",'"'))['name'])
    except:
        collection.append(None)
```

```python
genres = []
for ele in df["genres"]:
    try:
        if pd.isnull(ele):
            genres.append([])
        else:
            _ = json.loads(ele.replace("'",'"'))
            names = list(map(lambda item: item['name'], _))
            genres.append(names)
    except:
        genres.append([])
```

```python
df.drop(columns=["belongs_to_collection","genres"], axis=1, inplace=True)
df["collection"] = collection
df["genres"] = genres
```

```python
df_ = df[["original_title","release_date","collection","genres"]]
```

## Create a analysis agent

```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain.agents import AgentType
from langchain_openai import ChatOpenAI
import pandas as pd
import numpy as np
```

```python
llm = ChatOpenAI(
    model='gpt-4o-mini',
    verbose=True
)
```

```python
agent = create_pandas_dataframe_agent(
    llm,
    df = df_,
    verbose=False,
    allow_dangerous_code=True,
    agent_type=AgentType.OPENAI_FUNCTIONS   
)
print("Data Analysis Agent is ready. You can now ask questions about the data.")
```

```python
def ask_agent(question):
    """Function to ask questions to the agent and display the response"""
    response = agent.run({
        "input": question,
        "agent_scratchpad": f"Human: {question}\nAI: To answer this question, I need to use Python to analyze the dataframe. I'll use plotly library if the question needs any vizualization, I'll use the python_repl_ast tool.\n\nAction: python_repl_ast\nAction Input: ",
    })
    print(f"Question: {question}")
    print(f"Answer: {response}")
    print("---")
```

```python
ask_agent("what are the column names of in this dataset, show in tabular form")
```

```python
ask_agent("I want to see number of games released in latest 5 years")
```
